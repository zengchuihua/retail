```python
import pandas as pd
f = open('D:\\data\\附件1.csv',encoding='gbk')
data = pd.read_csv(f)  
print(data)
```

```python
data.drop(labels = range(70679,70680), axis = 0, inplace = True)
data=data.loc[(data['应付金额'] >=data['实际金额']) & data['应付金额'] > 0]
A=data.loc[data['地点'] == 'A']
B=data.loc[data['地点'] == 'B']
C=data.loc[data['地点'] == 'C']
D=data.loc[data['地点'] == 'D']
E=data.loc[data['地点'] == 'E']
data.to_csv('D:\\data\\onefinal.csv ',encoding='gbk',index=False)
A.to_csv('D:\\data\\task1-1A.csv ',encoding='gbk',index=False)
B.to_csv('D:\\data\\task1-1B.csv ',encoding='gbk',index=False)
C.to_csv('D:\\data\\task1-1C.csv ',encoding='gbk',index=False)
D.to_csv('D:\\data\\task1-1D.csv ',encoding='gbk',index=False)
E.to_csv('D:\\data\\task1-1E.csv ',encoding='gbk',index=False)
```

```python
data_file = ['D:\\data\\task1-1A.csv','D:\\data\\task1-1B.csv','D:\\data\\task1-1C.csv',
             'D:\\data\\task1-1D.csv','D:\\data\\task1-1E.csv']
month = ['2017/1/','2017/2/','2017/3/','2017/4/','2017/5/','2017/6/','2017/7/','2017/8/','2017/9/','2017/10/','2017/11/','2017/12/']
monday=[31,28,31,30,31,30,31,31,30,31,30,31]
i=0
for file in data_file:
    task = pd.read_csv(file,encoding='gbk')
    for mon in month:
        mon_data = task[task.支付时间.str.startswith(mon)]
        average_money = mon_data['实际金额'].sum() / mon_data.shape[0]
        print('{}售货机{}的每单平均交易额为：{}'.format(file[-5],mon,average_money))
        average_number = mon_data.shape[0] / monday[i]
        i+=1
        print('{}售货机{}的日均订单量为：{}单\n'.format(file[-5],mon,average_number))
    i=0
```
