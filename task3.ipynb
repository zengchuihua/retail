```python
import pandas as pd
import numpy as np
```

```python
data=pd.DataFrame(pd.read_csv('D:\\data\\task1-1A.csv',encoding='gbk'))
f = open('D:\\data\\附件2.csv',encoding='gbk')
data2= pd.read_csv(f)  #读取附件2
data['大类']='未知种类'
drinks=list(data2[data2.大类=='饮料'].商品)
for i in drinks:
    data[data.商品==i]=data[data.商品==i].replace('未知种类','饮料')
turnover=data.loc[data['大类'] == '饮料']
Group = turnover[['商品']].groupby(by='商品')
sale=Group.size()
df=pd.DataFrame(sale,columns=['订单量'])
df['标签']='未知'
df.loc[df.订单量>=220]=df.replace('未知','热销')
df.loc[(50<df.订单量)&(df.订单量<220)]=df.replace('未知','正常')
df.loc[df.订单量<=50]=df.replace('未知','滞销')
df=df.drop(columns=['订单量'])
df.to_csv('D:\\data\\task3-1A.csv',encoding='gbk')
```

```python
data=pd.DataFrame(pd.read_csv('D:\\data\\task1-1E.csv',encoding='gbk'))
data.商品.to_csv('D:\\data\\E.txt')
import matplotlib.pyplot as plt
import jieba
from wordcloud import WordCloud
#1.读出歌词
text = open('D:\\data\\E.txt','r',encoding='utf-8').read()
#2.把歌词剪开
cut_text = jieba.cut(text)
# print(type(cut_text))
#3.以空格拼接起来
result = " ".join(cut_text)
# print(result)
# 4.生成词云
wc = WordCloud(
    font_path='simhei.ttf',     #字体路劲
    background_color='white',   #背景颜色
    width=1000,
    height=600,
    max_font_size=90,            #字体大小
    min_font_size=15,
    mask=plt.imread('D:\\Google\\xinxin.jpg'),  #背景图片
    max_words=1000
)
wc.generate(result)
wc.to_file('D:\\data\\E.png')    #图片保存
#5.显示图片
plt.figure('jielun')   #图片显示的名字
plt.imshow(wc)
plt.axis('off')        #关闭坐标
plt.show()
```
