```python
#####task2-2售货机每月总交易额折线图#####
data_file=['D:\\data\\task1-1A.csv','D:\\data\\task1-1B.csv','D:\\data\\task1-1C.csv','D:\\data\\task1-1D.csv','D:\\data\\task1-1E.csv']
month = ['2017/1/','2017/2/','2017/3/','2017/4/','2017/5/','2017/6/','2017/7/','2017/8/','2017/9/','2017/10/','2017/11/','2017/12/']
temp = {}
for file in data_file:
    task = pd.read_csv(file,encoding='gbk')
    pla = []
    for mon in month:
        mon_data = task[task.支付时间.str.startswith(mon)]
        monsum=mon_data['实际金额'].sum()
        pla.append(monsum)
        print('{}售货机{}的每月总交易额为：{}'.format(file[-5],mon,monsum))
    temp[file[-5]] = pla
```

```python
dates = pd.date_range('2017/1',periods=12,freq='M')
values = pd.DataFrame(temp,index=dates)
pd.DataFrame(temp,index=dates)  #每台售货机每月总交易额数据
```

```python
for cl in class_two:
    for i in list(data2.loc[data2.二级类==cl].商品.values):
        data1.loc[data1.商品==i]=data1.loc[data1.商品==i].replace('未知类别',cl)
```

```python
data1.head()
```

```python
month = ['2017/1/','2017/2/','2017/3/','2017/4/','2017/5/','2017/6/','2017/7/','2017/8/','2017/9/','2017/10/','2017/11/','2017/12/']
monday=[31,28,31,30,31,30,31,31,30,31,30,31]
temp = {}
i=0
for cl in class_two:
    pla = []
    for mon in month:
        mon_data = data1[data1.支付时间.str.startswith(mon)]
        aaa=mon_data[(mon_data.二级类==cl)].实际金额.sum()/ monday[i]
        #data.loc[data['地点'] == 'A']
        pla.append(aaa)
        i+=1
        print('{}{}月份的交易额均值为{}'.format(cl,mon,aaa))
    temp[cl] = pla
    i=0
```

```python
######二级类目交易额均值#####
dates = pd.date_range('2017/1',periods=12,freq='M')
values = pd.DataFrame(temp,index=dates)
values
```
